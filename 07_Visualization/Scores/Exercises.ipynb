### Step 1. Import the necessary libraries

```python
from pyspark.sql import SparkSession, Row
import pyspark.sql.functions as F
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spark = SparkSession.builder.master("local[1]").appName("scores").getOrCreate()
```

### Step 2. Create the DataFrame that should look like the one below.

```python
df = spark.createDataFrame(
    [Row(first_name='Jason',last_name='Miller', age=42, female=0, preTestScore=4, postTestScore=25),
    Row(first_name='Molly',last_name='Jacobson', age=52, female=0, preTestScore=24, postTestScore=94),
    Row(first_name='Tina',last_name='Ali', age=36, female=0, preTestScore=31, postTestScore=57),
    Row(first_name='Jake',last_name='Milner', age=24, female=0, preTestScore=2, postTestScore=62),
    Row(first_name='Amy',last_name='Cooze', age=73, female=0, preTestScore=3, postTestScore=70)]
)
```

```python
df.show()
```

```python
df.printSchema()
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
df_pd = df.toPandas()
```

```python
sns.scatterplot(y=df_pd['preTestScore'], x=df_pd['postTestScore'], size=df_pd['age'])
```

```python
sns.scatterplot(y=df_pd['postTestScore'], x=df_pd['preTestScore'], size=df_pd['age'])
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sns.scatterplot(y=df_pd['postTestScore']*4.5, x=df_pd['preTestScore'], size=df_pd['age'])
```
